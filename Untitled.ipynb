```python
import matplotlib.pyplot as plt
import numpy as np

theta = np.linspace(0, 2*np.pi, 100)

a= np.sqrt(20.0)
b= np.sqrt(15.0)
c= np.sqrt(10.0)
d= np.sqrt(5.0)

a1 = a*np.cos(theta)
a2 = a*np.sin(theta)
b1 = b*np.cos(theta)
b2 = b*np.sin(theta)
c1 = c*np.cos(theta)
c2 = c*np.sin(theta)
d1 = d*np.cos(theta)
d2 = d*np.sin(theta)

fig, ax = plt.subplots(1)

ax.plot(a1, a2, 'b')
ax.plot(b1, b2,'b--')
ax.plot(c1, c2,'b*')
ax.plot(d1, d2, 'y')

plt.xlim(-4.95,4.95)
plt.ylim(-4.95,4.95)



plt.show()
```
